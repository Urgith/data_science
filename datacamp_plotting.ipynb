```python
# 01.09.2025
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel('datacamp.xlsx')
data = data.dropna(how='all')
data = data.reset_index(drop=True)
```

```python
new_df = pd.DataFrame(data.iloc[:, 0])
new_df['Mean Score'] = data.iloc[:, 1:].mean(axis=1)

new_df = new_df.sort_values('Mean Score', ascending=False)
```

```python
plt.bar(new_df['Course Name'], new_df['Mean Score'])
plt.hlines(y=min(new_df['Mean Score']), xmin=0, xmax=56.5, color='red')
plt.annotate(xy=(56, min(new_df['Mean Score'])), xytext=(53, 3), text=min(new_df['Mean Score']), color='red', fontsize=20)

plt.grid()
plt.xticks(rotation=90, fontsize=6)

plt.xlabel('Course Name')
plt.ylabel('Mean Score')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data
```

```python
data.info()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.grid()
```

```python
sns.scatterplot(x='size', y='total_bill', data=data)
plt.grid()
```

```python
sns.scatterplot(x='size', y='tip', data=data)
plt.grid()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=data)
plt.grid()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=data)
plt.grid()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='day', data=data)
plt.grid()
```

```python
sns.set_style('darkgrid')

y = []
for row in df.index:
    y.append(df.loc[row, :].mean())

plt.plot(y)
```

Średnie wartości dla wierszy wahają się między -20 a +20.


```python
print(df[df['1'] < 0]['1'].mean())
print(df[df['1'] < -10]['1'].mean())
print(df[df['1'] < -20]['1'].mean())
print(df[df['1'] < -30]['1'].mean())
print(df[df['1'] < -50]['1'].mean())
```

Średnia gdy bierzemy pod uwagę wartości mniejsze od 0 jest zawsze wyższa, niż średnia, gdy bierzemy pod uwagę wartości mniejsze od liczb ujemnych.

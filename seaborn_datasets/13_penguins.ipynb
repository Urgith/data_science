```python
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
```

```python
df.info()
```

```python
df = df.dropna()
df
```

```python
df['species'].value_counts().plot(kind='pie')
df['species'].value_counts()
```

Najwięcej, bo aż 146 jest ptaków gatunku 'Adelie', następnie 'Gentoo' z 119 wystąpieniami, a na szarym końcu 'Chinstrap' liczące 68 okazów.


```python
df['island'].value_counts().plot(kind='pie')
df['island'].value_counts()
```

Jeżeli chodzi o wyspy występowania, to najwięcej jest z 'Biscoe' aż 163 ptaki pochodzą stamtąd,
następnie jest wyspa 'Dream' z 123 okazami, a ostatnia wyspa 'Torgersen' tylko 47 ptaków.


```python
sns.barplot(x='species', y='bill_length_mm', data=df)
```

Na powyższym rysunku widać, że gatunek 'Chinstrap' ma najdłuższy dziób, natomiast 'Adelie', najkrótszy dziób.


```python
sns.barplot(x='species', y='bill_depth_mm', data=df)
```

Jeżeli chodzi o głębokość dzioba, to najpłytszy jest dziób ptaka 'Gentoo', a 2 pozostałe ptaki mają dzioby o podobnej głębokości.


```python
sns.barplot(x='species', y='flipper_length_mm', data=df)
```

Jeżeli chodzi o długość płetw, to najdłuższą płetwę ma ptak 'Gentoo', następny jest 'Chinstrap', a na samym końcu mamy 'Adelie'


```python
sns.barplot(x='species', y='body_mass_g', data=df)
```

Ptak 'Gentoo' jest zdecydowanie najbardziej masywny, dwa pozostałe gatunki mają podobne masy.


```python
sns.barplot(x='sex', y='body_mass_g', data=df)
```

Bez zaskoczeń samce ptaków omawianych gatunków są cięższe od samic.


```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df)
```

Na powyższym wykresie widać, że głębokość dziobu idzie w parze z jego długością. Rozbicie danych na gatunki ptaków uwidacznia tą zależność jeszcze bardziej.


```python
sns.regplot(x='flipper_length_mm', y='body_mass_g', data=df)
```

Im dłuższe skrzydła ptaka, tym większa jego masa.

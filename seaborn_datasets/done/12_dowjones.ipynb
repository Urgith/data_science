```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('dowjones')
```

```python
df.info()
```

```python
df
```

```python
sns.set_style('darkgrid')
```

```python
sns.lineplot(x='Date', y='Price', data=df)
```

Akcje spółki 'dowjones' od roku 1932 do 1970 miały tendencje wzrostostowe.


```python
df['pct_change'] = df['Price'].pct_change()
```

```python
sns.scatterplot(x='Date', y='pct_change', data=df)
```

Na wykresie powyżej widoczne są zwroty akcji spółki na przestrzeni 56 lat.


```python
sns.histplot(x='pct_change', data=df)
plt.vlines(0, 0, 80, color='r')
plt.vlines(df['pct_change'].median(), 0, 80, color='g')
```

Większa część zwrotów wartości akcji znajduje się na prawo od 0, czyli spółka w tym przediale czasowym zyskała na wartości.
Mediana (zielona kreska) jest również dodatnia.

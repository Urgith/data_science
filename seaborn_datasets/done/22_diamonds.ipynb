```python
df['volume'] = df['x']  * df['y'] * df['z']
```

```python
import numpy as np
```

```python
model = np.poly1d(np.polyfit(df['carat'], df['price'], 2))
y = np.linspace(0, 3, 1000)
```

```python
sns.scatterplot(x='carat', y='price', data=df)
sns.lineplot(x=y, y=model(y))
```

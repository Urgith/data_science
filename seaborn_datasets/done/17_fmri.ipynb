```python
import seaborn as sns
```

```python
df = sns.load_dataset('fmri')
```

```python
df.info()
```

```python
df
```

```python
sns.set_style('darkgrid')
```

```python
sns.lineplot(df.groupby('timepoint')['signal'].agg('mean'))
```

Na wykresie powyżej widać, że śednia wygnału w określonym momencie czasu waha się w granicach od około -0.06 do 0.135.
 

```python
sns.lineplot(df.groupby('subject')['signal'].agg('mean'))
```

'subject' nie ma zbyt dużego wpływu no moc sygnału.

```python
import seaborn as sns
```

```python
df = sns.load_dataset('planets')
```

```python
df.info()
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
plt.xticks(rotation=90)

sns.countplot(x='method', data=df)
```

Istnieją 2 najczęstrze metody znajdowania planet: 'Radial Velocity' oraz 'Transit'.


```python
sns.countplot(y='number', data=df)
```

Na powyższym wykresie widać, że najczęściej są odkrywane plaety pojedynczo, następnie pod dwie, trzy, cztery itd. aż do 7.


```python
sns.histplot(x='mass', data=df)
plt.vlines(df.mass.mean(), 0, 200, color='red')
plt.vlines(df.mass.median(), 0, 200, color='green')
```

Jak widać na wykresie wyżej, najwięcej jest planet o niskiej masie.


```python
sns.histplot(x='distance', data=df)
plt.vlines(df.distance.mean(), 0, 250, color='red')
plt.vlines(df.distance.median(), 0, 250, color='green')
```

Podobnie z odległością planet, najwięcej jest tych, które są najbliżej.


Zarówno masa jak i odległość planet są prawostronnie skośne.


```python
sns.boxplot(x='orbital_period', data=df)
plt.xscale('log')
```

Okres orbitalny planet jest bardzo rozproszony, waha się od mniej niż 1 aż do ponad 700 tyś.


```python
sns.violinplot(y='year', data=df)
```

Na wykresie powyżej widać, że najwięcej planet odkryto w roku 2011. Pierwsze planety wykrywano już w latach 80'.

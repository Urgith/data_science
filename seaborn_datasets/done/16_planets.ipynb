```python
import seaborn as sns
```

```python
df = sns.load_dataset('planets')
```

```python
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
plt.xticks(rotation=90)

sns.countplot(x='method', data=df)
```

There are two main planet searching methods: Radial Velocity and Transit.


```python
sns.countplot(y='number', data=df)
```

```python
sns.histplot(x='mass', data=df)
```

Jak widać na wykresie wyżej, najwięcej jest planet o niskiej masie.


```python
sns.histplot(x='distance', data=df)
```

Podobnie z odległością planet, najwięcej jest tych, które są najbliżej


```python
sns.boxplot(x='orbital_period', data=df)
plt.xscale('log')
```

Okres orbitalny plaet jest bardzo rozproszony, waha się od mniej niż 1 aż do ponad 700 tyś.

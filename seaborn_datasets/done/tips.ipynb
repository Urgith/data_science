```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
```

Na powyszym wykresie wida, e napiwek jest wikszy, gdy wicej pacono za jedzenie w restauracji.


```python
sns.barplot(x='sex', y='tip', data=df)
```

Mczyni maj tendencj do dawania wyszych napiwkw.


```python
sns.barplot(x='smoker', y='tip', data=df)
```

Informacja o tym, e klient pali papierosy, nie ma wikszego wpywu na to jak wysoki da napiwek.


```python
sns.barplot(x='day', y='tip', data=df)
```

Napiwki s wysze w weekend ni w czwartek i pitek.


```python
sns.barplot(x='time', y='tip', data=df)
```

Klienci daj wiksze napiwki w porze kolacji ni w porze obiadu.

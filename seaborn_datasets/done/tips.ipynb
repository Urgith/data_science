```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
```

Na powyższym wykresie widać, że napiwek jest większy, gdy więcej płacono za jedzenie w restauracji.


```python
sns.barplot(x='sex', y='tip', data=df)
```

Mężczyźni mają tendencję do dawania wyższych napiwków.


```python
sns.barplot(x='smoker', y='tip', data=df)
```

Informacja o tym, że klient pali papierosy, nie ma większego wpływu na to jak wysoki da napiwek.


```python
sns.barplot(x='day', y='tip', data=df)
```

Napiwki są wyższe w weekend niż w czwartek i piątek.


```python
sns.barplot(x='time', y='tip', data=df)
```

Klienci dają większe napiwki w porze kolacji niż w porze obiadu.

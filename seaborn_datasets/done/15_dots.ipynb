```python
df['firing_rate'].plot(kind='hist')
```

Większość 'firing_rate' mieści się w przedziale od 30 do 60.


```python
df.plot(x='coherence', y='firing_rate', kind='scatter')
```

```python
by_align = df.groupby(by='align').firing_rate.mean()
by_align
```

Średnie dla align='dots' oraz align='sacc' są prawie identyczne.


```python
by_choice = df.groupby(by='choice').firing_rate.mean()
by_choice
```

Z koleji średnia dla 'choice'=T1 jest wyższa niż dla 'choice'=T2.


```python
sns.scatterplot(x='time', y='firing_rate', hue='choice', data=df)
```

```python
sns.scatterplot(x='time', y='firing_rate', hue='align', data=df)
```

2 Wykresy powyżej są dosyć specyficzne, pokazują zależności między 'time', 'firing_rate', 'choice' oraz 'align'.


```python
sns.scatterplot(x='time', y='coherence', hue='align', data=df)
```

Wykres powyżej pokazuje, że przedział czasu się zmniejsza wraz ze wzrostem koherentności. Widać również, że align='dots' występuje dopiero wtedy, kiedy 'time' >= 300.

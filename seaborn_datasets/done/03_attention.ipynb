```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('attention')
```

```python
df.info()
```

```python
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
sns.set_style('darkgrid')
```

```python
for solution in range(1, 4):
    data = df[df['solutions'] == solution]
    sns.lineplot(x='subject', y='score', data=data, label=solution)

    print(data.groupby('solutions').score.mean())

plt.xticks(range(1, 21))
plt.legend()
```

Na wykresie jak rwnie i z oblicze wynika, e rednio najmniej punktw jest dla 1 rozwizania (5.35), rednia ilo punktw dla dwch rozwiza (5.975), a najwiksza ilos punktw dla 3 rozwiza (6.55).


```python
for attention in ('divided', 'focused'):
    data = df[df['attention'] == attention]
    sns.lineplot(x=range(1, 31), y='score', data=data, label=attention)

plt.xticks(range(1, 33, 2));
```

Na wykresie poniej wida, e dla skupionej uwagi, wynik oscyluje w granicach od 5 do 9. Natomiast, gdy uwaga jest podzielona, to wynik ronie majc na pocztku wartoci w granicha 3, a pniej w granicach 7.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('attention')
```

```python
df.info()
```

Wszystkie dane są kompletne.


```python
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

Niepotrzebna kolumna 'Unnamed: 0' została usunięta.


```python
sns.set_style('darkgrid')
```

```python
for solution in range(1, 4):
    data = df[df['solutions'] == solution]
    sns.lineplot(x='subject', y='score', data=data, label=solution)

    print(data.groupby('solutions').score.mean())

plt.xticks(range(1, 21))
plt.legend()
```

Na wykresie jak również i z obliczeń wynika, że średnio najmniej punktów jest dla 1 rozwiązania (5.35), średnia ilość punktów dla dwóch rozwiązań (5.975), a największa ilosć punktów dla 3 rozwiązań (6.55).


```python
for attention in ('divided', 'focused'):
    data = df[df['attention'] == attention]
    sns.lineplot(x=range(1, 31), y='score', data=data, label=attention)

    print(data.groupby('attention').score.mean())

plt.xticks(range(1, 33, 2));
```

Na wykresie poniżej widać, że dla skupionej uwagi, wynik oscyluje w granicach od 5 do 9. Natomiast, gdy uwaga jest podzielona, to wynik rośnie mając na początku wartości w granicha 3, a później w granicach 7.
Poza tym średni wynik dla skupionej uwagi to 6.8 a dla podzielonej 5.12.


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.attention_focused = (df.attention_focused == True).astype(int)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('score', axis=1), df['score'])
```

```python
model = LinearRegression()
model.fit(X_train, y_train);
```

```python
model.score(X_test, y_test)
```

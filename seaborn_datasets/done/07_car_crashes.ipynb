```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.set_style('darkgrid')
```

```python
df.total.plot(kind='hist')
plt.xlabel('Total car crashes')
```

```python
df.speeding.plot(kind='hist')
plt.xlabel('Speeding car crashes')
```

```python
df.alcohol.plot(kind='hist')
plt.xlabel('Alcohol car crashes')
```

```python
df.not_distracted.plot(kind='hist')
plt.xlabel('Facused on driving')
```

```python
df.no_previous.plot(kind='hist')
plt.xlabel('No previous car crashes')
```

```python
df.ins_premium.plot(kind='hist')
plt.xlabel('Insurance')
```

```python
df.ins_losses.plot(kind='hist')
plt.xlabel('Lost insurance')
```

```python
sns.heatmap(df.iloc[:, :-1].corr())
```

```python
sns.swarmplot(data=df.iloc[:, :-3])
```

Na powyższym wykresie widzimy, że większość wypadków powodują kierowcy skupieni na jeżdzie. Nie byli oni również uczestnikami wypadków drogowych. Spożycie alkoholu oraz prekraczanie prędkości stanowią niewielką część wypadków.


```python
sns.barplot(y='abbrev', x='total', data=df.sort_values(by='total', ascending=False))
plt.yticks(size=6);
```

Do największej liczby wypadków na liczbę mieszczńców dochodzi w Północnej Karolinie. Z koleii najmniej wypadków zdarza się w Dystrykcie Kolumbii.

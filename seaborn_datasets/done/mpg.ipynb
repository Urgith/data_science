```python
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
sns.histplot(data=df, x='mpg')
```

```python
sns.pairplot(data=df[:-2])
```

```python
sns.scatterplot(x='mpg', y='weight', hue='horsepower', data=df)
```

Z wykresu powyżej widać, że im cięższy samochód, tym mniej mil może przejechać na 1 galonie benzyny.


Poza tym, można zauważyć, że im mniej koni mechanicznych ma samochód, tym więcej może przejechać na 1 galonie benzyny.

```python
# 19.07.2025
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
mean = df['Score'].mean()
df = df[:20]
df
```

```python
plt.bar(x=df['Team'], height=df['Score'])
plt.axhline(mean, color='red')

plt.title('Volleyball TOP20')
plt.ylabel('Score')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

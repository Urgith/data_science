```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
plt.plot(df['Level'], df['Experience'])

plt.title('Tibia total_exp nedded to reach Level')
plt.ylabel('Experience')
plt.xlabel('Level')

plt.ticklabel_format(style='plain')
plt.grid()
plt.show()
```

```python
df['Diff'] = df['Experience'].diff()
```

```python
plt.plot(df['Level'], df['Diff'])

plt.title('Exp nedded to gain 1 Level')
plt.ylabel('Experience')
plt.xlabel('Level')

plt.ticklabel_format(style='plain')
plt.grid()
plt.show()
```

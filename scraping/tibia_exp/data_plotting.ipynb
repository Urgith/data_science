```python
# 19.07.2025
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
poly = np.polyfit(df['Level'], df['Experience'], 3)
poly
```

```python
ax = df.plot('Level', 'Experience')
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ','))
)
x = df['Level']
plt.plot(x, x**3*poly[0] + x**2*poly[1] + x*poly[2] + poly[3])

plt.title('Tibia total_exp nedded to reach Level')
plt.ylabel('Experience')
plt.xlabel('Level')

plt.grid()
plt.show()
```

```python
df['Diff_Experience'] = df['Experience'].diff()
```

```python
poly = np.polyfit(df['Level'][1:], df['Diff_Experience'][1:], 2)
poly
```

```python
ax = df.plot('Level', 'Diff_Experience')
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ','))
)
x = df['Level'][1:]
plt.plot(x, x**2*poly[0] + x*poly[1] + poly[2])

plt.title('Exp nedded to gain 1 Level')
plt.ylabel('Experience')
plt.xlabel('Level')

plt.grid()
plt.show()
```

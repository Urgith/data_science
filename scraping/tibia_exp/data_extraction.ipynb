```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
full_data = []

with open('data_table.txt', encoding='utf-8') as file:
    text = file.read()

soup = BeautifulSoup(text, 'html.parser')
table = soup.find('tr', {'class': 'FlexTable'})
rows = table.find_all('td')[3:]
```

```python
new_rows = []
for row in rows:
    new_row = row.text.replace(',', '')

    if new_row.isnumeric():
        new_rows.append(new_row)

level = []
exp = []
for i, row in enumerate(new_rows):
    if i % 2 == 0:
        level.append(int(row))
    else:
        exp.append(int(row))

print(level)
print(exp)
```

```python
df = pd.DataFrame({'Level': level, 'Experience': exp})
df
```

```python
df.to_excel('data.xlsx', index=False)
```

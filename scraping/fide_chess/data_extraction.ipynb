```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open('chess_ranking.txt', 'r', encoding='utf-8') as file:
    text = file.read()
```

```python
soup = BeautifulSoup(text, 'html.parser')
rows = soup.find('table').find_all('tr')
```

```python
full_data = []
for column in rows[1:]:
    new_col = [col for col in column.text.split(' ') if col]

    full_row = []
    for entry in new_col:
        entry = [col for col in entry.split('\n') if col]
        full_row.extend(entry)

    length = len(full_row) - 6
    new_name = ' '.join(full_row[1:3 + length])

    for _ in range(length + 1):
        full_row.pop(1)

    full_row[1] = new_name
    full_data.append(full_row)
```

```python
df = pd.DataFrame(full_data)
df
```

```python
df.columns = ['#', 'Name', 'Fed', 'Rating', 'B-Year']
df['#'] = df['#'].astype(int)
df['Rating'] = df['Rating'].astype(int)
df['B-Year'] = df['B-Year'].astype(int)
```

```python
df.to_excel('data.xlsx', index=False)
```

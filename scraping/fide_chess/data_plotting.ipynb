```python
# 18.07.2025
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
grouped = df.groupby('Fed')['Rating']
new_df = grouped.max().sort_values(ascending=True)
```

```python
new_df.plot.barh()

plt.grid(axis='x')
plt.xlim([2600, 2900])
plt.title('Highest rated player by country')
plt.ylabel('Country')
plt.xlabel('Rating')

plt.show()
```

```python
new_df = df.groupby('B-Year')['Rating']
grouped = new_df.mean()
grouped.index = 2025 - grouped.index
```

```python
grouped.plot.bar()
plt.ylim([2600, 2800])
plt.title('Average rating by age')
plt.ylabel('Mean rating')
plt.xlabel('Age')

plt.grid(axis='y')
plt.show()
```

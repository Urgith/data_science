```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
mean = df['Score'].mean()
var = df['Score'].var()
```

```python
plt.hist(df['Score'], bins=100)
plt.axvline(mean, color='red')
plt.annotate(text=f'Mean: {mean.round(4)}', xy=(6.5, 625), color='red', size=15)

plt.title('Scores distribution on MyAnimeList')
plt.ylabel('Number of scores')
plt.xlabel('Mean score')
plt.grid()
plt.show()
```

```python
df_tv = df[df['Type'] == 'TV']

mean = df_tv['Score'].mean()
var = df_tv['Score'].var()
```

```python
plt.hist(df_tv['Score'], bins=100)
plt.axvline(mean, color='red')
plt.annotate(text=f'Mean: {mean.round(4)}', xy=(7.5, 190), color='red', size=15)

plt.title('TV Scores distribution on MyAnimeList')
plt.ylabel('Number of scores')
plt.xlabel('Mean score')
plt.grid()
plt.show()
```

```python
df = df[df['Type'] != 'Unknown']

sns.pointplot(data=df, x='Type', y='Score', linewidth=2)

plt.title('Score by Anime Type')
plt.ylabel('Mean Score')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
sns.boxplot(data=df, x='Type', y='Score', hue='Type')

plt.title('Score by Anime Type')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

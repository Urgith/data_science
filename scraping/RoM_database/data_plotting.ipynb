```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df = df.iloc[:, 1:]
```

```python
sns.set_style('darkgrid')
sns.relplot(data=df, x='level', y='HP', row='typ', hue='typ', palette='tab10')

plt.yscale('log')
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
full_data = []

for i, file in enumerate([f'data/RoM_database{number}.txt' for number in range(1, 343)]):
    with open(file, 'r', encoding='utf-8') as file_text:
        text = file_text.read()

    soup = BeautifulSoup(text, 'html.parser')
    row = soup.find('table').find_all('tr')

    for element in row:
        datas = element.find_all('td')
        full_data.append([data.text for data in datas])

full_data
```

```python
columns = full_data[0]
columns
```

```python
for monster in full_data[1:]:
    monster[1] = monster[1].split('(')[0]
    monster[2] = monster[2][0]
    monster[3] = monster[3].split('+-')[0]
    monster[4] = monster[4].split('(')[0]
    monster[5] = monster[5].split()
    monster[6] = monster[6].split()
    monster[8] = monster[8].split()
```

```python
data_indices = []
for i, element in enumerate(full_data):
    if i % 31 == 0:
        data_indices.append(i)

data_indices.reverse()

for index in data_indices:
    full_data.pop(index)

full_data
```

```python
HP = []
MP = []
DMG = []
pAtk = []
mAtk = []
pCritRes = []
mCritRes = []
pdef = []
mdef = []

EP = []
TP = []

for monster in full_data:
    monster[4] = monster[4].strip()

    for i, attribute in enumerate(monster[5]):
        attribute = attribute.split('HP:')[0]
        attribute = attribute.split('MP:')[0]
        attribute = attribute.split('DMG:')[0]
        attribute = attribute.split('pAtk:')[0]
        attribute = attribute.split('mAtk:')[0]
        attribute = attribute.split('pCritRes')[0]
        attribute = attribute.split('mCritRes:')[0]
        attribute = attribute.split('pdef:')[0]
        attribute = attribute.split('mdef:')[0]

        if i == 1:
            HP.append(''.join(attribute.split('.')))
        elif i == 2:
            MP.append(''.join(attribute.split('.')))
        elif i == 3:
            DMG.append(''.join(attribute.split('.')))
        elif i == 4:
            pAtk.append(''.join(attribute.split('.')))
        elif i == 5:
            mAtk.append(''.join(attribute.split('.')))
        elif i == 6:
            pCritRes.append(''.join(attribute.split('.')))
        elif i == 7:
            mCritRes.append(''.join(attribute.split('.')))
        elif i == 8:
            pdef.append(''.join(attribute.split('.')))
        elif i == 9:
            mdef.append(''.join(attribute.split('.')))

    for i, attribute in enumerate(monster[8]):
        attribute = attribute.split('EP:')[0]
        attribute = attribute.split('TP:')[0]

        if i == 1:
            EP.append(''.join(attribute.split('.')))
        elif i == 2:
            TP.append(''.join(attribute.split('.')))
```

```python
for i, monster in enumerate(full_data):
    monster.extend([HP[i], MP[i], DMG[i], pAtk[i], mAtk[i], pCritRes[i], mCritRes[i], pdef[i], mdef[i]])
    monster.extend([EP[i], TP[i]])
    monster.pop(9)
    monster.pop(8)
    monster.pop(6)
    monster.pop(5)
    monster.pop(0)
```

```python
columns.extend(['HP', 'MP', 'DMG', 'pAtk', 'mAtk', 'pCritRes', 'mCritRes', 'pdef', 'mdef', 'EP', 'TP'])
columns.remove('attribute')
```

```python
columns.remove('guid')
columns.remove('loot')
columns.remove('xp/tp')
columns.remove('raisechange')
```

```python
df = pd.DataFrame(full_data, columns=columns)
df
```

```python
splitted_respawn = df.respawn.str.split(':')

elements = []
for element in splitted_respawn:
    if element == ['-1']:
        elements.append(-1)
    else:
        elements.append(int(element[0])*3600 + int(element[1])*60 + int(element[2]))

df['respawn_seconds'] = elements
df = df.drop('respawn', axis=1)
```

```python
df.info()
```

```python
df.typ = df.typ.astype(np.int64)
df.level = df.level.astype(np.int64)
df.HP = df.HP.astype(np.int64)
df.MP = df.MP.astype(np.int64)
df.DMG = df.DMG.astype(np.int64)
df.pAtk = df.pAtk.astype(np.int64)
df.mAtk = df.mAtk.astype(np.int64)
df.pCritRes = df.pCritRes.astype(np.int64)
df.mCritRes = df.mCritRes.astype(np.int64)
df.pdef = df.pdef.astype(np.int64)
df.mdef = df.mdef.astype(np.int64)
df.EP = df.EP.astype(np.int64)
df.TP = df.TP.astype(np.int64)
```

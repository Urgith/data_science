```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data.xlsx')
```

```python
df = df.iloc[:, 1:]
df_size = df.shape[1]
```

```python
df = df.sort_values(f'ranking_{df_size // 2}')
```

```python
df = df.iloc[:, 0:df_size:2]
```

```python
sns.set_style('darkgrid')

for row in df.iterrows():
    plt.plot(range(1, df_size // 2 + 1), row[1].values[1:])

plt.xticks(range(1, df_size // 2 + 1, 2));
```

```python
df.iloc[:, 1:] = df.iloc[:, 1:].cumsum(axis=1) / np.arange(1, df_size // 2 + 1)
```

```python
for row in df.iterrows():
    plt.plot(range(1, df_size // 2 + 1), row[1].values[1:])

plt.xticks(range(1, df_size // 2 + 1, 2));
```

```python
# 18.07.2025
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import squarify
```

```python
mpl.style.use('classic')
```

```python
df = pd.read_excel('data.xlsx')
```

```python
new_df = df['Country'].value_counts()
```

```python
squarify.plot(sizes=new_df, label=new_df.index, text_kwargs={'rotation': 30, 'fontsize': 10})

plt.title('Państwa z największą liczbą firm w TOP100')

plt.axis(False)
plt.show()
```

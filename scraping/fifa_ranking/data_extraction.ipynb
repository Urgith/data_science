```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open('data_fifa.txt', 'r', encoding='utf-8') as file:
    text = file.read()
```

```python
soup = BeautifulSoup(text, 'html.parser')
table = soup.find('table', {'role': 'table'}).find_all('td')[:-8]
```

```python
country = []
code = []
total_points = []
previous_points = []
difference = []
```

```python
for j, entry in enumerate(table):
    data = entry.text

    if j % 7 == 1:
        country.append(data[:-3])
        code.append(data[-3:])
    elif j % 7 == 2:
        total_points.append(data)
    elif j % 7 == 3:
        previous_points.append(data)
    elif j % 7 == 4:
        difference.append(data)
```

```python
df = pd.DataFrame([country, code, total_points, previous_points, difference]).transpose()
df['Rank'] = df.index + 1
df.columns = ['Country', 'Code', 'Total_Points', 'Previous_Points', 'Difference', 'Rank']
df = df[['Rank', 'Country', 'Code', 'Total_Points', 'Previous_Points', 'Difference']]
df
```

```python
df['Total_Points'] = df['Total_Points'].astype(float)
df['Previous_Points'] = df['Previous_Points'].astype(float)

df['Difference'] = df['Difference'].str.replace('-', '0')
df['Difference'] = df['Difference'].astype(float)
```

```python
df.to_excel('data.xlsx', index=False)
```

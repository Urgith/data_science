```python
# 18.07.2025
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
```

```python
new_df = df.sort_values('Previous_Points', ascending=False)
new_df.index = range(1, 211)
new_df = new_df.reset_index()
```

```python
new_df['Difference'] = new_df['Total_Points'] - new_df['Previous_Points']
```

```python
new_df.columns = ['Previous_Rank', *new_df.columns[1:]]
new_df = new_df[['Rank', 'Previous_Rank', *new_df.columns[2:]]]
new_df = new_df.sort_values('Rank')
```

```python
new_df['Rank_Difference'] = new_df['Previous_Rank'] - new_df['Rank']
```

```python
plt.style.use('ggplot')

plt.scatter(new_df['Difference'], new_df['Rank_Difference'])
plt.title('FIFA Rating Diff VS Points Diff')
plt.ylabel('Rating change')
plt.xlabel('Rank change')
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df = df[['Country', 'Population (2025)', 'Land Area (Km^2)', 'Density (P/Km^2)']]
df['Population (2025)'] = df['Population (2025)'].str.replace(',', '').astype(int)
df['Land Area (Km^2)'] = df['Land Area (Km^2)'].str.replace(',', '').astype(int)
df = df.sort_values('Land Area (Km^2)', ascending=False)
df = df.reset_index(drop=True)[7:]
```

```python
ax = sns.scatterplot(data=df, x='Land Area (Km^2)', y='Population (2025)', hue='Density (P/Km^2)', legend=False)
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.get_xaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.title('Land Area VS Population')
plt.xticks(rotation=45)

plt.grid()
plt.show()
```

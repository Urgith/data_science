```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df = df[['Country', 'Population (2025)', 'Land Area (Km^2)']]
df['Population (2025)'] = df['Population (2025)'].str.replace(',', '').astype(int)
df['Land Area (Km^2)'] = df['Land Area (Km^2)'].str.replace(',', '').astype(int)
df = df.sort_values('Land Area (Km^2)', ascending=False)
df = df.reset_index(drop=True)[7:]
```

```python
df.plot(kind='scatter', x='Land Area (Km^2)', y='Population (2025)')

plt.ticklabel_format(style='plain')
plt.title('Land Area VS Population')
plt.xticks(rotation=45)

plt.grid()
plt.show()
```

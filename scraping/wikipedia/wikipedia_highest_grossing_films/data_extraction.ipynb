```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
with open('data_box_office.txt', 'r', encoding='utf-8') as file:
    text = file.read()
```

```python
soup = BeautifulSoup(text, 'html.parser')
table = soup.find('table', {'class': 'wikitable sortable plainrowheaders sticky-header col4right col5center col6center'})
```

```python
body = table.find('tbody')
```

```python
data = body.find_all('tr')
```

```python
full_data = []
```

```python
for element in data:
    full_data.append([ele for ele in element.text.split('\n')[:-2] if ele])
```

```python
df = pd.DataFrame(full_data)
df = df.drop(0)
```

```python
df.columns = ['Rank', 'Peak', 'Title', 'Gross', 'Year']
```

```python
df['Peak'] = df['Peak'].str.replace('RK', '')
df['Peak'] = df['Peak'].str.replace('TS3', '')

df['Gross'] = df['Gross'].str.replace('T', '')
df['Gross'] = df['Gross'].str.replace('F', '')
df['Gross'] = df['Gross'].str.replace('8$', '')
df['Gross'] = df['Gross'].str.replace('SM', '')
df['Gross'] = df['Gross'].str.replace('DKR', '')
df['Gross'] = df['Gross'].str.replace('$', '')
```

```python
df.Rank = df.Rank.astype(int)
df.Peak = df.Peak.astype(int)

df.Gross = df.Gross.str.replace(',', '')
df.Gross = df.Gross.astype(np.int64)

df.Year = df.Year.astype(int)

df
```

```python
df.to_excel('data.xlsx', index=False)
```

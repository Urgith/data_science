```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
grouped = df.groupby('Year')['Gross'].count()
grouped = grouped.reindex(range(grouped.index[0], grouped.index[-1] + 1), fill_value=0)
```

```python
plt.plot(grouped, marker='o')

plt.title('Number of movies by Year')
plt.ylabel('Number of Movies')
plt.xlabel('Year')

plt.grid()
plt.show()
```

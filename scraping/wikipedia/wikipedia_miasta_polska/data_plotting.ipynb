```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
```

```python
grouped = df.groupby('Województwo')['Liczba mieszkańców'].sum().sort_values()
```

```python
plt.barh(y=grouped.index, width=grouped)

plt.title('Liczba mieszkańców w miastach > 100 000')
plt.xlabel('Liczba Mieszkańców')
plt.ylabel('Województwo')

plt.ticklabel_format(style='plain', axis='x')
plt.grid(axis='x')
plt.show()
```

```python
# 19.07.2025
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
```

```python
grouped = df.groupby('Województwo')['Liczba mieszkańców'].sum().sort_values()
```

```python
plt.barh(y=grouped.index, width=grouped)

plt.title('Liczba mieszkańców w miastach > 100 000')
plt.xlabel('Liczba Mieszkańców')
plt.ylabel('Województwo')

plt.ticklabel_format(style='plain', axis='x')
plt.grid(axis='x')
plt.show()
```

```python
ax = df.plot.bar(x='Miasto', y='Liczba mieszkańców', legend=False)
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ','))
)

plt.title('Największe miasta w Polsce')
plt.ylabel('Populacja')
plt.grid(axis='y')
plt.show()
```

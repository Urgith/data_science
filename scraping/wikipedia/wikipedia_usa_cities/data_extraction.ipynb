```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open('data.txt', encoding='utf-8') as file:
    text = file.read()

soup = BeautifulSoup(text, 'html.parser')
table = soup.find('table', {'style': 'font-size:88%;margin-left: auto; margin-right: auto; width: 100%; border: 1px solid darkgray;'})
```

```python
data = table.find_all('td')

rank = []
name = []
region = []
population = []

for i, element in enumerate(data):
    if i % 8 == 0 or i % 8 == 4:
        rank.append(element.text)

    elif i % 8 == 1 or i % 8 == 5:
        name.append(element.text)

    elif i % 8 == 2 or i % 8 == 6:
        region.append(element.text)

    elif i % 8 == 3 or i % 8 == 7:
        population.append(element.text)
```

```python
df = pd.DataFrame([rank, name, region, population]).transpose()
df.columns = ['Rank', 'Name', 'Region', 'Population']
```

```python
df.Rank = df.Rank.astype(int)

df.Population = df.Population.str.replace(',', '')
df.Population = df.Population.astype(int)
```

```python

df = df.sort_values('Rank')
df
```

```python
df.to_excel('data.xlsx', index=False)
```

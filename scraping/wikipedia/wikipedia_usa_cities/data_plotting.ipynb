```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
```

```python
sns.barplot(data=df, x='Name', y='Population', hue='Region')

plt.title('Biggest US Cities')
plt.xlabel('City')

plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

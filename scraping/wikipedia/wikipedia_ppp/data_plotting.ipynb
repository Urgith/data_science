```python
# 20.07.2025
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df.PPP = df.PPP * 1_000_000
```

```python
others = df.iloc[22:, 1].sum()
```

```python
df = df.drop(range(22, 223), axis=0)
df = df.drop(0, axis=0)
```

```python
df.iloc[-1] = ['Other', others]
```

```python
plt.style.use('classic')

ax = df.plot.bar(x='Country', y='PPP', legend=False)
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ','))
)

plt.xticks(rotation=90)
plt.grid(axis='y', linewidth=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df.PPP = df.PPP * 1_000_000
```

```python
others = df.iloc[22:, 1].sum()
```

```python
df = df.drop(range(22, 223), axis=0)
df = df.drop(0, axis=0)
```

```python
df.iloc[-1] = ['Other', others]
```

```python
plt.bar(x=df['Country'], height=df['PPP'])

plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

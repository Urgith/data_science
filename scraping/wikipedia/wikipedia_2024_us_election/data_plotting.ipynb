```python
# 19.07.2025
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df = df.sort_values('Total votes')[:-1]
```

```python
slice = df[-20:]
slice['Republican Won'] = slice['Trump/Vance Republican Votes'] > slice['Harris/Walz Democratic Votes']
slice['Republican Won'] = slice['Republican Won'].map({True: 'red', False: 'blue'})
```

```python
slice.plot.barh(y='Total votes', x='State', color=slice['Republican Won'])

plt.ticklabel_format(style='plain', axis='x')
plt.xticks(rotation=45)
plt.legend().remove()
plt.grid(axis='x')

plt.title('20 Top States by Number of Votes')
plt.xlabel('Number of Votes')
plt.show()
```

```python
trump = df['Trump/Vance Republican EV'].sum()
harris = df['Harris/Walz Democratic EV'].sum()
```

```python
plt.barh(y='Trump', width=trump, label='Trump')
plt.barh(y='Trump', width=harris, left=trump, label='Harris')
plt.axvline(270, color='red')

plt.legend()
plt.show()
```

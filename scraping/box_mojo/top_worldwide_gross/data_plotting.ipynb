```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
grouped = df.groupby('Year')['Worldwide_gross']
new_df = grouped.sum()
```

```python
new_df.plot()

plt.ticklabel_format(style='plain')
plt.title('Worldwide Box Office by Year')
plt.ylabel('Box Office')
plt.grid()

plt.annotate(f'MAX: {max(new_df)}', xy=(2019, max(new_df)), xytext=(1975, 27500000000),
             arrowprops={'facecolor': 'black', 'shrink': 0.05})
plt.annotate(f'MIN: {new_df[2020]}', xy=(2020, new_df[2020]), xytext=(1965, 7500000000),
             arrowprops={'facecolor': 'black', 'shrink': 0.05})
```

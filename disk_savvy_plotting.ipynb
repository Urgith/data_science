```python
# 01.09.2025
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel('disk_savvy.xlsx')
data
```

```python
plt.plot(data.date, data.disk_space)

plt.xticks(rotation=45)
plt.grid()

plt.title('Python Packages Storage by Time')
plt.xlabel('Date')
plt.ylabel('MB of Storage')

plt.show()
```

```python
plt.plot(data.date, data.files)

plt.xticks(rotation=45)
plt.grid()

plt.title('Python Packages Files')
plt.xlabel('Date')
plt.ylabel('Number of Files')

plt.show()
```
